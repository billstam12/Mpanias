```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
gamma=10**8               #παραγοντας lorentz αρχικου e
Es=1.6022*10**(-10)         #Ενεργεια πεδίου φωτονιων απο τον δισκο σε erg   
G=[gamma]
E1= np.linspace(0.000000001, 1, 50,endpoint=True)        #υπολογισμός απωλειών στο klein-Nishina
def q(E1,Ge):
    return E1/(1+Ge*(1-E1))
def apwl(Ge,y):
    return (math.log10(Ge)**(-1))*(1+0.5*((1-y)*(Ge*y)**2)/(1+Ge*y))
def Ge(gamma):
    return (4*Es*gamma)/(8.18712228*10**(-7))
def calc_loss(gamma):
    ge = Ge(gamma)
    y=q(E1,ge)
    z=apwl(ge,y)
    max_z = max(z)
    hta = [E1[z.argmax()]]
    return hta[0]

gen=0                                       #υπολογισμός multiplicity 
hta = [calc_loss(gamma)]
while max(G)>3.81*10**3:
    gen=gen+1
    for i in range(0,len(G),1):
        if G[i]>3.81*10**3:
            G.append(hta[i]/2*G[i])
            hta1 = calc_loss(gamma)
            hta.append(hta1)
            G.append(hta[i]/2*G[i])
            hta2 = calc_loss(gamma)
            hta.append(hta2)
            G[i]=(1-hta[i])*G[i]
            hta[i] = calc_loss(G[i])
print('Total electrons:',len(G))            #ποσα e δημιουργηθηκαν(υποθετωντας μονο αντιστροφο compton)
Gnew=np.unique(G)                           #τελικοι παραγοντες lorentz τελικων ηλεκτρονιων
Gnew
```

```python
j=0                                   #πλήθος ηλεκτρονίων για κάθε ξεχωριστό παραγοντα lorentz
su=[0]
for i in range(0,len(G)-1,1):
    if G[i] == G[i+1]:
        su[j] = su[j] + 1
    else :
        su.append(0)
        j = j+1
for i in range(0,len(su),1):
    su[i]=su[i]+1
su  
```

```python
plt.figure(figsize=(20,10))
plt.ylim(ymax = 3.5, ymin = 0)

fEph=[]
num=[]
for k in Gnew:
    l=0
    i=0
    gamma=k
    E=su[l]*gamma*8.2*10**(-7)
    Ee=[E]
    Eph=[su[l]*(2.14*10**(-10))*gamma**2]
    while gamma > 1:
        Ee.append((Ee[i]-Eph[i]))
        gamma = Ee[i+1]/(su[l]*8.2*10**(-7))
        Eph.append(su[l]*2.14*10**(-10)*gamma**2)
        i=i+1
    fEph.append(Eph)
    l+=1
    print len(Eph)
    x=[y for y in range(0,len(Eph))]
    plt.scatter(np.log10(x),Eph)
```
